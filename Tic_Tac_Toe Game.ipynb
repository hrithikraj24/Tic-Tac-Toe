```python
board = {'1':' ', '2':' ', '3':' ',
         '4':' ', '5':' ', '6':' ',
         '7':' ', '8':' ', '9':' '}

board_keys=[]

for key in board:
    board_keys.append(key)
    
    
def printboard(board):
    print(board['1']+'|'+board['2']+'|'+board['3'])
    print('-+-+-')
    print(board['4']+'|'+board['5']+'|'+board['6'])
    print('-+-+-')
    print(board['7']+'|'+board['8']+'|'+board['9'])
    
def game():
    turn = 'X'
    c=0
    for i in range(10):
        printboard(board)
        print("Your turn now! "+turn)
        move=input()
        
        if(board[move]==' '):
            board[move]=turn
            c=c+1
        else:
            print('The place is already filled. Move to someplace else!')
            continue
        
        if(c>=5):
            if(board['1'] == board['2'] == board['3'] != ' '):
                print("Game Over!!")
                print("Player "+turn+ " wins!!")
                printboard(board)
                break
            elif(board['4'] == board['5'] == board['6'] != ' '):
                print("Game Over!!")
                print("Player "+turn+ " wins!!")
                printboard(board)
                break
            elif(board['7'] == board['8'] == board['9'] != ' '):
                print("Game Over!!")
                print("Player "+turn+ " wins!!")
                printboard(board)
                break
            elif(board['1'] == board['4'] == board['7'] != ' '):
                print("Game Over!!")
                print("Player "+turn+ " wins!!")
                printboard(board)
                break
            elif(board['2'] == board['5'] == board['8'] != ' '):
                print("Game Over!!")
                print("Player "+turn+ " wins!!")
                printboard(board)
                break
            elif(board['3'] == board['6'] == board['9'] != ' '):
                print("Game Over!!")
                print("Player "+turn+ " wins!!")
                printboard(board)
                break
            elif(board['1'] == board['5'] == board['9'] != ' '):
                print("Game Over!!")
                print("Player "+turn+ " wins!!")
                printboard(board)
                break
            elif(board['3'] == board['5'] == board['7'] != ' '):
                print("Game Over!!")
                print("Player "+turn+ " wins!!")
                printboard(board)
                break
            
        if(c==9):
            print("The match is tied!!")
            printboard(board)
            
        if(turn == 'X'):
            turn = '0'
        else:
            turn ='X'
            
    restart = input("Do you wish to restart the game?(Y/N)")
    if(restart == 'Y' or restart=='y'):
        for key in board_keys:
            board[key]=' '
                
        game()
        
if __name__=="__main__":
    game()
```
